# 1 : Common room types in data set

```python
##  Common room types in data set

# Count the occurrences of each unique value in the 'roomType' column
room_type_counts = df['roomType'].value_counts()

# Increase figure size
plt.figure(figsize=(20, 6))

# Create a bar chart
plt.bar(room_type_counts.index, room_type_counts.values)

# Add labels and title
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Count of Each Unique Room Type')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=90)

# Display the plot
plt.show()


# Display the plot
plt.tight_layout()  # Ensures labels fit within the figure area
plt.show()

```

```python
## Taking only fewer room types

# Count the occurrences of each unique value in the 'roomType' column
room_type_counts = df['roomType'].value_counts()

# Select the top 20 room types
top_room_types = room_type_counts.head(20)

# Increase figure size
plt.figure(figsize=(20, 6))

# Create a bar chart
plt.bar(top_room_types.index, top_room_types.values)

# Add labels and title
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Count of Top 20 Room Types')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()  # Ensures labels fit within the figure area
plt.show()
```

# 2 :  Prices varying across room types

```python
## Prices varying across room types

# Create a box plot or violin plot
plt.figure(figsize=(20, 6))  # Set figure size
sns.set(style="whitegrid")  # Set plot style

# Choose either box plot or violin plot by uncommenting the appropriate line
#sns.boxplot(x="roomType", y="price", data=df)  # Box plot
sns.violinplot(x="roomType", y="price", data=df)  # Violin plot

# Add labels and title
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.title('Comparison of Prices for Each Room Type')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

# 3 :  Avg price for each city

```python
df.columns
```

```python
# Group data by City and calculate the average price
average_prices = df.groupby("city")["price"].mean().reset_index()

# Create a bar chart using matplotlib
plt.figure(figsize=(20, 6))
plt.bar(average_prices["city"], average_prices["price"], color='skyblue')
plt.xlabel("city")
plt.ylabel("Average Price")
plt.title("Average Prices for Each City (Bar Chart)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Create a box plot using seaborn
plt.figure(figsize=(20, 6))
sns.boxplot(x="city", y="price", data=df, palette="Set3")
plt.xlabel("City")
plt.ylabel("Price")
plt.title("Distribution of Prices by City (Box Plot)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 4 :  Host response time vs price

```python
# Group data by responseTime and calculate the average price
average_prices = df.groupby("responseTime")["price"].mean().reset_index()

# Create a bar chart using matplotlib
plt.figure(figsize=(10, 6))
plt.bar(average_prices["responseTime"], average_prices["price"], color='skyblue')
plt.xlabel("Response Time")
plt.ylabel("Average Price")
plt.title("Average Prices for Each Response Time (Bar Chart)")
plt.xticks(average_prices["responseTime"])
plt.tight_layout()
plt.show()

# Create a box plot using seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(x="responseTime", y="price", data=df, palette="Set3")
plt.xlabel("Response Time")
plt.ylabel("Price")
plt.title("Distribution of Prices by Response Time (Box Plot)")
plt.tight_layout()
plt.show()
```

# 5 : Number of guests frequency

```python
# Create a histogram using matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['numberOfGuests'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Number of Guests")
plt.ylabel("Frequency")
plt.title("Histogram of Number of Guests")
plt.tight_layout()
plt.show()

# Create a bar chart using matplotlib
guest_counts = df['numberOfGuests'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(guest_counts.index, guest_counts.values, color='skyblue')
plt.xlabel("Number of Guests")
plt.ylabel("Frequency")
plt.title("Bar Chart of Number of Guests")
plt.xticks(guest_counts.index)
plt.tight_layout()
plt.show()
```

# 6 : Price vs review counts

```python
# Create a box plot to identify outliers in the 'price' column
plt.figure(figsize=(8, 6))
plt.boxplot(df['price'], vert=False)
plt.xlabel("Price")
plt.title("Box Plot of Price to Identify Outliers")
plt.tight_layout()
plt.show()

# Create a box plot to identify outliers in the 'Review Count' column
plt.figure(figsize=(8, 6))
plt.boxplot(df['Review Count'], vert=False)
plt.xlabel("Review Count")
plt.title("Box Plot of Review Count to Identify Outliers")
plt.tight_layout()
plt.show()

# Create a scatter plot to visualize the relationship between 'price' and 'Review Count'
plt.figure(figsize=(8, 6))
plt.scatter(df['price'], df['Review Count'], color='skyblue')
plt.xlabel("Price")
plt.ylabel("Review Count")
plt.title("Scatter Plot of Price vs. Review Count")
plt.tight_layout()
plt.show()
```

# 7 : MaxNight vs MinNight

```python
# Create a scatter plot of 'price' against 'maxNights'
plt.figure(figsize=(8, 6))
plt.scatter(df['maxNights'], df['price'], color='skyblue')
plt.xlabel("Max Nights")
plt.ylabel("Price")
plt.title("Scatter Plot of Price vs. Max Nights")
plt.tight_layout()
plt.show()

# Create a scatter plot of 'price' against 'minNights'
plt.figure(figsize=(8, 6))
plt.scatter(df['minNights'], df['price'], color='skyblue')
plt.xlabel("Min Nights")
plt.ylabel("Price")
plt.title("Scatter Plot of Price vs. Min Nights")
plt.tight_layout()
plt.show()

# Create a line plot of 'price' against 'maxNights'
plt.figure(figsize=(8, 6))
plt.plot(df['maxNights'], df['price'], marker='o')
plt.xlabel("Max Nights")
plt.ylabel("Price")
plt.title("Line Plot of Price vs. Max Nights")
plt.tight_layout()
plt.show()

# Create a line plot of 'price' against 'minNights'
plt.figure(figsize=(8, 6))
plt.plot(df['minNights'], df['price'], marker='o')
plt.xlabel("Min Nights")
plt.ylabel("Price")
plt.title("Line Plot of Price vs. Min Nights")
plt.tight_layout()
plt.show()
```

# 8 : Person capacity vs Number of bedrooms

```python

# Create a scatter plot of 'personCapacity' against 'numberOfBedsAvailable'
plt.figure(figsize=(8, 6))
plt.scatter(df['personCapacity'], df['numberOfBedsAvailable'], color='skyblue')
plt.xlabel("Person Capacity")
plt.ylabel("Number of Beds Available")
plt.title("Scatter Plot: Person Capacity vs. Number of Beds Available")
plt.tight_layout()
plt.show()

# Group data by 'personCapacity' and calculate the average 'numberOfBedsAvailable'
average_beds = df.groupby("personCapacity")["numberOfBedsAvailable"].mean().reset_index()

# Create a bar chart to compare average 'numberOfBedsAvailable' by 'personCapacity'
plt.figure(figsize=(10, 6))
plt.bar(average_beds["personCapacity"], average_beds["numberOfBedsAvailable"], color='skyblue')
plt.xlabel("Person Capacity")
plt.ylabel("Average Number of Beds Available")
plt.title("Bar Chart: Average Number of Beds Available by Person Capacity")
plt.xticks(average_beds["personCapacity"])
plt.tight_layout()
plt.show()
```
