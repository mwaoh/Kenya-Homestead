```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("datafiles/cleaned_air_bnb_Jul_26.csv")
print(df.shape)
df.head()
```

```python
# folow data_exploration_vicky.ipynb file
df = df.drop(columns=['url','name','address','localizedCheckInTimeWindow','localizedCheckOutTime','responseTime',''],axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

average_prices = df.groupby(["bathroomLabel"],sort = True)["price"].mean().reset_index()

# Create a bar chart using matplotlib
plt.figure(figsize=(20, 6))
plt.bar(average_prices["bathroomLabel"], average_prices["price"], color='skyblue')
plt.xlabel("bathroomLabel")
plt.ylabel("Average Price")
plt.title("Average Prices for Each bathroomLabel (Bar Chart)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
